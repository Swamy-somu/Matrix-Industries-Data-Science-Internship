```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)

print("="*60)
print("TASK 2: DATA VISUALIZATION & EDA")
print("="*60)

url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

print(f"\n✓ Dataset loaded: {df.shape[0]} rows, {df.shape[1]} columns")
print(f"\nFirst 5 rows:")
print(df.head())

# 1. Survival Count
print("\n✓ Chart 1: Overall Survival Count")
plt.figure(figsize=(8, 5))
sns.countplot(x="Survived", data=df, palette="Set1")
plt.title("Overall Survival Count (0=Died, 1=Survived)", fontsize=14, fontweight="bold")
plt.xlabel("Survival Status")
plt.ylabel("Count")
plt.show()

# 2. Survival by Sex
print("\n✓ Chart 2: Survival by Sex")
plt.figure(figsize=(8, 5))
sns.countplot(x="Sex", hue="Survived", data=df, palette="husl")
plt.title("Survival by Sex", fontsize=14, fontweight="bold")
plt.xlabel("Sex")
plt.ylabel("Count")
plt.show()
print("Insight: Women had significantly higher survival rates.")

# 3. Survival by Passenger Class
print("\n✓ Chart 3: Survival by Passenger Class")
plt.figure(figsize=(8, 5))
sns.countplot(x="Pclass", hue="Survived", data=df, palette="cool")
plt.title("Survival by Passenger Class", fontsize=14, fontweight="bold")
plt.xlabel("Passenger Class")
plt.ylabel("Count")
plt.show()
print("Insight: 1st class passengers had much higher survival rates.")

# 4. Age Distribution
print("\n✓ Chart 4: Age Distribution")
plt.figure(figsize=(10, 5))
sns.histplot(df["Age"], bins=30, kde=True, color="skyblue")
plt.title("Age Distribution of Passengers", fontsize=14, fontweight="bold")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()

# 5. Age vs Survival
print("\n✓ Chart 5: Age vs Survival")
plt.figure(figsize=(10, 5))
sns.boxplot(x="Survived", y="Age", data=df, palette="Set2")
plt.title("Age Distribution by Survival Status", fontsize=14, fontweight="bold")
plt.xlabel("Survived (0=No, 1=Yes)")
plt.ylabel("Age")
plt.show()

# 6. Fare Distribution
print("\n✓ Chart 6: Fare Distribution")
plt.figure(figsize=(10, 5))
sns.histplot(df["Fare"], bins=40, kde=True, color="lightcoral")
plt.title("Fare Distribution", fontsize=14, fontweight="bold")
plt.xlabel("Fare ($)")
plt.ylabel("Frequency")
plt.show()

# 7. Fare vs Class
print("\n✓ Chart 7: Fare by Passenger Class")
plt.figure(figsize=(10, 5))
sns.boxplot(x="Pclass", y="Fare", data=df, palette="muted")
plt.title("Fare Distribution by Passenger Class", fontsize=14, fontweight="bold")
plt.xlabel("Passenger Class")
plt.ylabel("Fare ($)")
plt.show()

# 8. Embarked Port
print("\n✓ Chart 8: Survival by Embarkation Port")
plt.figure(figsize=(10, 5))
sns.countplot(x="Embarked", hue="Survived", data=df, palette="rocket")
plt.title("Survival by Embarkation Port", fontsize=14, fontweight="bold")
plt.xlabel("Port")
plt.ylabel("Count")
plt.show()

# 9. Correlation Heatmap
print("\n✓ Chart 9: Correlation Heatmap")
plt.figure(figsize=(10, 8))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f", square=True)
plt.title("Correlation Matrix of Numeric Features", fontsize=14, fontweight="bold")
plt.show()

print("\n" + "="*60)
print("KEY INSIGHTS SUMMARY")
print("="*60)
print("✓ Women Survival: Women had significantly higher survival rates")
print("✓ Class Effect: 1st class passengers survived at much higher rates")
print("✓ Fare Correlation: Higher fares correlated with higher survival")
print("✓ Age Factor: Younger passengers had slightly better survival")
print("\n✓ TASK 2 COMPLETED! All visualizations generated.")
print(f"Total passengers: {len(df)}")
print(f"Survived: {df['Survived'].sum()} ({df['Survived'].mean()*100:.1f}%)")
print(f"Died: {(1-df['Survived']).sum()} ({(1-df['Survived'].mean())*100:.1f}%)")
```
